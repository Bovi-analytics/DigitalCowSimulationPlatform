```python
from pandas import read_parquet
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
footprint = read_parquet("../memfootprint.parquet")
footprint = (footprint / 1024 ** 2).round(2)
footprint
```

```python
f, ax = plt.subplots(figsize=(5.5, 4.5))
sns.barplot(footprint, ax=ax, palette="flare")
ax.set_yscale("log")
ax.set_title("SciPy sparse matrix size in multiple formats")
ax.set_xlabel("Sparse format")
ax.set_ylabel("Matrix size (mebibytes)")
plt.show()
```
